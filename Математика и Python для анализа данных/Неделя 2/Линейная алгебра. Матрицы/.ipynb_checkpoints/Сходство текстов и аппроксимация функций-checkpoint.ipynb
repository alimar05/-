```python
import re
import numpy as np
from itertools import chain
from scipy.spatial.distance import cosine

with open("sentences.txt", 'r') as file_obj:
    list_sentences = list(map(str.lower, file_obj.readlines()))

n_sentences = len(list_sentences)
list_tokenize_sentences = [list(filter(lambda word: word not in [''],
                                       re.split('[^a-z]', list_sentences[i])))
                           for i in range(n_sentences)]
dict_words_sentences = {word : idx for idx, word in enumerate(set(chain(*list_tokenize_sentences)))}

matrix_word_frequency_sentences = np.zeros(shape=(n_sentences, len(dict_words_sentences)))

for i in range(n_sentences):
    for word in list_tokenize_sentences[i]:
        matrix_word_frequency_sentences[i][dict_words_sentences[word]] += 1

idxs_two_similar_sentences = np.argsort([cosine(matrix_word_frequency_sentences[0][:],
                                                matrix_word_frequency_sentences[i][:])
                                         for i in range(n_sentences)])[1:3]

with open("Задача 1.txt", 'w') as file_obj:
    file_obj.write(' '.join(list(map(str, idxs_two_similar_sentences))))
```

```python
from math import sin, exp
from matplotlib import pylab as plt
%matplotlib inline

def f(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

# Область определения x и y соответственно
D = np.arange(1, 15, 0.1)
E = list(map(f, D))
plt.plot(D, E)

x = np.array([1, 15])
A = np.array([[1, x[0]],
              [1, x[1]]])
b = np.array([f(x[0]),
              f(x[1])])
w = np.linalg.solve(A, b)
y = A.dot(w)
plt.plot(x, y)

x = np.array([1, 8, 15])
A = np.array([[1, x[0], x[0]**2],
              [1, x[1], x[1]**2],
              [1, x[2], x[2]**2]])
b = np.array([f(x[0]),
              f(x[1]),
              f(x[2])])
w = np.linalg.solve(A, b)
y = A.dot(w)
plt.plot(x, y)

x = np.array([1, 4, 10, 15])
A = np.array([[1, x[0], x[0]**2, x[0]**3],
              [1, x[1], x[1]**2, x[1]**3],
              [1, x[2], x[2]**2, x[2]**3],
              [1, x[3], x[3]**2, x[3]**3]])
b = np.array([f(x[0]),
              f(x[1]),
              f(x[2]),
              f(x[3])])
w = np.linalg.solve(A, b)
y = A.dot(w)
plt.plot(x, y)

with open('Задача 2.txt', 'w') as file_obj:
    file_obj.write(' '.join(map(str, np.round(w, 2))))
```
