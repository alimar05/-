# Экспоненциальное распределение

```python
import numpy as np
from scipy.stats import expon, norm, gaussian_kde
import matplotlib.pyplot as plt
%matplotlib inline

SIZE = 1000

# Зададим экспоненциально распределённую непрерывную случайную величину
scale = 500
rv_expon = expon(scale=scale)
# Возьмём выборку из этой случайной величины объёмом 1000
sample = rv_expon.rvs(size=SIZE)
```

Гистограмма выборки:

```python
# Оформление графика
font = {'size': 10}
plt.rc('font', **font)

# density=True для того чтобы площадь под гистограммой была равна 1
plt.hist(sample, density=True, alpha=0.5, label='Empirical')
# Теоретическая плотность распределения случайной величины
x = np.linspace(np.min(sample), np.max(sample), SIZE)
plt.plot(x, rv_expon.pdf(x), label='Theoretical', c='r')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

Оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. Для этого при трёх и более значениях $n$ (например, 5, 10, 50) сгенерируйте 1000 выборок объёма $n$ и постройте гистограммы распределений их выборочных средних.

Используя информацию о среднем и дисперсии исходного распределения, посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних.

```python
# Каждому ключу n в thousand_sample_mean_n соответствует 1000 выборочных средних из n
n_sizes = [1, 5, 10, 30, 50, 100]
thousand_sample_mean_n = {n: [np.mean(rv_expon.rvs(size=n)) for _ in range(SIZE)]
                          for n in n_sizes}
```

```python
# Оформление графиков
font = {'size': 25}
plt.rc('font', **font)
figsize = (30, 20)

# Определим матожидание и дисперсию исходного распределения
lambd = 1.0 / scale
EX = 1.0 / lambd
DX = (1.0 / lambd) ** 2

hist = {}
n_rows = 2
n_cols = 3
fig, axes = plt.subplots(figsize = figsize, nrows=n_rows, ncols=n_cols)
for i in range(n_rows):
    for j in range(n_cols):
        hist[n_cols * i + j] = axes[i][j].hist(thousand_sample_mean_n[n_sizes[n_cols * i + j]],
                                               density=True, label='empiric')
        # Аппроксимация нормальным распределением
        x = np.linspace(np.min(thousand_sample_mean_n[n_sizes[0]]),
                        np.max(thousand_sample_mean_n[n_sizes[0]]), SIZE)
        rv_norm = norm(loc=EX, scale=np.sqrt(DX / n_sizes[n_cols * i + j]))
        axes[i][j].plot(x, rv_norm.pdf(x), label='norm', c='r', lw=5)
        axes[i][j].ticklabel_format(axis='y', style='sci', scilimits=(0, 0))
        axes[i][j].set_title('$n$ = ' + str(n_sizes[n_cols * i + j]))
        axes[i][j].set_ylabel('$f(x)$')
        axes[i][j].set_xlabel('$x$')
        axes[i][j].legend(loc='best')
```
